# Machine Learning Preprocessing

Sistem rekomendasi yang akan kita buat adalah Content-Based Filtering. Artinya, sistem akan merekomendasikan produk berdasarkan kemiripan konten (deskripsi, tipe produk, kategori, dan kandungan utama/ key ingredients).

Di level ini, kita gabungkan semua informasi teks itu menjadi satu kolom fitur yang disebut tags.

```python
# Memilih kolom-kolom yang relevan untuk sistem rekomendasi
features = ['Product Name', 'Brand', 'Product Type', 'Description', 'Key_Ingredients', 'Main_Category']

# Membuat copy dataframe khusus untuk sistem rekomendasi
df_rec = df_products[features].copy()

# Mengisi nilai yang kosong (NaN) dengan spasi agar tidak error saat diproses
df_rec.fillna('', inplace=True)

# Membuat kolom baru bernama 'tags'
# Kolom ini adalah gabungan dari Kategori, Tipe, Ingredients, dan Deskripsi
df_rec['tags'] = df_rec['Main_Category'] + " " + df_rec['Product Type'] + " " + df_rec['Key_Ingredients'] + " " + df_rec['Description']

print("Data setelah digabungkan menjadi tags:")
display(df_rec[['Product Name', 'tags']].head(3))
```

# Level 1: The Kantan (Simple) Tier – Market Clustering

Goal: Discover hidden market segments using Unsupervised Learning.

*   The Algorithm: K-Means Clustering or Principal Component Analysis (PCA)
*   The Data: We use products_cleaned.csv, focusing strictly on numerical and binary columns: safety risk percentages (low/moderate/high) and formulation traits (alcohol_free, fragrance_free, paraben_free).
*  The Output: The model will automatically group products into distinct clusters—for example, it might independently discover a cluster of "Ultra-Safe Sensitive Skin Basics" versus "High-Active Chemical Treatments." It's a fantastic visual way to see where local Indonesian brands sit compared to global ones!



To perform K-Means Clustering, our machine learning model needs the data to be perfectly clean and purely mathematical. K-Means is a "distance-based" algorithm, meaning it groups products by calculating the physical distance between their data points.

Because of this, we have to do two critical preparation steps:

Feature Selection: Strip away all the text (like names and descriptions) and keep only numbers and binary flags (0s and 1s).

Standard Scaling: Balance the weights of our data. A safety ratio might range from 0.0 to 1.0, while a boolean flag is exactly 0 or 1. Scaling ensures K-Means doesn't prioritize one feature over another just because the raw numbers look different.

__Here is the Python code you need to prepare the data. You can run this in your local Jupyter Notebook, or you can upload the products_cleaned.csv file here in the chat and I will run it for you!

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# 1. Load the dataset
# (Assuming the file is in the same directory as your script)
df = pd.read_csv('products_cleaned.csv')

# Let's peek at the data to ensure it loaded correctly
print("Original Dataset Shape:", df.shape)

# 2. Select our specific features for clustering
# We are grabbing the formulation claims and safety risk columns
# (Column names might need a slight adjustment based on the exact CSV headers)
clustering_features = [
    'alcohol_free',
    'fragrance_free',
    'paraben_free',
    'sulfate_free',
    'silicone_free',
    'low_risk_ratio',       # Replace with actual column name if different
    'moderate_risk_ratio',  # Replace with actual column name if different
    'high_risk_ratio'       # Replace with actual column name if different
]

# Create a new dataframe with just our clustering features
X = df[clustering_features].copy()

# 3. Handle Missing Values
# If a product is missing a trait flag, we can safely assume it's a 0 (False).
X = X.fillna(0)

# 4. Scale the Data (Crucial for K-Means!)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Data is scaled and ready for clustering! Shape:", X_scaled.shape)
```

# Level 2: The Choudo Ii (Intermediate) Tier – NLP Classification

Goal: Teach a model to understand skincare routines.

*   The Algorithm: TF-IDF (Term Frequency-Inverse Document Frequency) paired with a classification model like Random Forest or Support Vector Machine (SVM).

*   The Data: The raw text inside the "How to Use" column.

*   The Output: We train the model to automatically classify products into categories like "Day vs. Night" or "Rinse-off vs. Leave-on." This introduces you to Natural Language Processing (NLP) text cleaning: removing stop words, stemming, and turning sentences into numbers.

Level 3: Membuat Sistem Rekomendasi (The Recommendation System!)
Sekarang kita masuk ke ranah Machine Learning! Kita akan menggunakan teknik TF-IDF (untuk mengubah teks di kolom tags menjadi vektor angka) dan Cosine Similarity (untuk menghitung skor kemiripan antar produk).

```python
#indo machine learning TF-IDF

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 1. Mengubah teks (tags) menjadi matriks angka menggunakan TF-IDF
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df_rec['tags'])

print(f"Ukuran Matriks TF-IDF: {tfidf_matrix.shape}")
print("(Artinya: Ada sekian produk dan sekian kata unik di dalam dataset kita)\n")

# 2. Menghitung kemiripan antar produk (Cosine Similarity)
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# 3. Membuat index untuk mempermudah pencarian produk berdasarkan nama
indices = pd.Series(df_rec.index, index=df_rec['Product Name']).drop_duplicates()

# 4. Membuat fungsi Sistem Rekomendasi
def get_recommendations(product_name, cosine_sim=cosine_sim):
    # Cek apakah produk ada di dataset
    if product_name not in indices:
        return "Waduh, produknya nggak ketemu di dataset nih."

    # Ambil index dari produk yang diinput
    idx = indices[product_name]

    # Ambil skor kemiripan produk tersebut dengan SEMUA produk lain
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Urutkan produk berdasarkan skor kemiripan tertinggi
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Ambil 5 produk teratas (index 0 pasti produk itu sendiri, jadi kita ambil 1-6)
    sim_scores = sim_scores[1:6]

    # Ambil index produk-produk hasil rekomendasi
    product_indices = [i[0] for i in sim_scores]

    # Tampilkan hasilnya dengan kolom yang rapi
    print(f"🔥 Karena kamu tertarik dengan '{product_name}', senpai rekomendasikan produk ini:")
    return df_products[['Product Name', 'Brand', 'Product Type', 'Key_Ingredients']].iloc[product_indices]

# ==========================================
# MARI KITA TEST SISTEM REKOMENDASINYA!
# ==========================================
# Kita coba ambil satu nama produk secara acak dari dataset (misal data di baris ke-10)
sample_product_to_test = df_products['Product Name'].iloc[10]

# Panggil fungsinya
display(get_recommendations(sample_product_to_test))
```
